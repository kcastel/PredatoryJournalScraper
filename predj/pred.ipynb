# Cleaning Article Information Scraped from Predatory Journals 
***

```python
import pandas as pd
```

```python
import re
def regexremove(abstract):
    regex = re.compile(r'[\n\r\t]')
    cleaned_abstract = regex.sub('', abstract)
    return(cleaned_abstract)

#https://stackoverflow.com/questions/9662346/python-code-to-remove-html-tags-from-a-string
def cleanhtml(raw_html):
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, '', raw_html)
    return cleantext

def cleaner(df):
    df.dropna(subset=['abstract'], inplace=True)
    df.reset_index(drop=True,inplace=True)
    df['abstract'] = df['abstract'].apply(cleanhtml)
    df['abstract'] = df['abstract'].apply(regexremove)
    df['abstract'] = df['abstract'].str.slice(start=8)
    return(df)
```

```python
rjeb = pd.read_csv('seventh.csv')
rjeb = cleaner(rjeb)

#Some abstracts start with 'abstract,' others don't -- need conditional filtering
rjebmod = rjeb[rjeb['abstract'].str.startswith('Abstract') == True]
rjebmod2 = rjeb[rjeb['abstract'].str.startswith('ABSTRACT') == True]

rjebmod = pd.concat([rjebmod,rjebmod2],axis=0)

rjebmod['abstract'] = rjebmod['abstract'].str.slice(start=8)
rjebmod = rjebmod.add_prefix('M_')

rjeb = pd.concat([rjeb,rjebmod],axis=1)
rjeb['abstract'] = rjeb['M_abstract'].fillna(rjeb['abstract'])
rjeb.drop(['M_abstract','M_author','M_title'],inplace=True,axis=1)

rjeb['Journal'] = 'Research Journal of Economics and Business Studies'
```
